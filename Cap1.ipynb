#Google Colaboratory
O Google Colaboratory, ou Google Colab, é um serviço de nuvem
que permite misturar códigos fonte, geralmente em python, e  linhas de texto, geralmente em ***markdown***. No próprio ambiente,  pode-se adiconar imagens, textos e gerar os resultados dos códigos. Se trata de uma técnica conhecida como ***notebook***.

Os ***notebooks*** são editores que permitem a construção, edição e execução de códigos. Tais programas podem ser executados em pequenas partes, conhecidas como **células**, o que facilita a visualização das informações. Além do mais, os notebooks podem ser compartilhados ou modificados por outros usuários, desde que tenham acesso liberado.

Inspirado no Jupyter Notebook, o Colab foi criado para ser uma ferramenta gratuita utilizada em ciência de dados. Para começar a utilizar o Colab, o usuário pode encontrar informações iniciais no *notebook* disponibilizado em [Welcome to Colab!](https://colab.research.google.com/notebooks/intro.ipynb)




#Criando o Primeiro Notebook

Logado no própria conta do Google, o usuário deve acessar o link anterior, que direciona para o *notebook* inicial. Na sequência, basta clicar em  >**File**, canto superior esquerdo, e em seguida em  **>New Notebook in Drive**, de acordo com a figura a seguir.

Fig1_PA_2024.png
          

O primeiro *notebook*, denominado *Untitled0.ipynb*, é criado. Na figura a seguir são destadas algumas funcionalidades do *notebook*.
Fig2_PA_2024.png


**1** **- Célula**: local utilizado para escrever códigos ou textos. Para executar o comando qualquer basta clicar no botão *play* ou então, utilizar o atalho no teclado SHIFT+ENTER.

**2** **+ Código**: adiciona células de código e
          **+ Texto**: adiciona as células de texto com as configurações markdown.

 **3** **- Estrutura de direrórios**: ao clicar no ícone da pasta aparecem opções como a de carregar arquivos do Google Drive* (*Montar Drive*) ou da própria máquina.
